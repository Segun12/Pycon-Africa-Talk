## Libraries used for webscraping
* BeautifulSoup: is a python library for parsing HTML and XML document
* Pandas: python library for data manipulation and analysis
* Requests: is a popular alternative to the standard library's HTTP handling (e.g. urllib2)

```python
# library importation
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get('https://steemit.com/nigeria/@leopantro/50-yoruba-proverbs-and-idioms')
soup = BeautifulSoup(res.content,'html.parser')
```

```python
# extracting necessary informations
yoruba = []
english = []
for ol in soup.findAll('ol'):
    proverb = ol.find('li')
    yoruba.append(proverb.text)

for paragraph in soup.findAll('p'):
    if paragraph.text.startswith("Meaning"):
        english.append(paragraph.text.split(':')[1].strip())
```

```python
df = pd.DataFrame({'Yoruba':yoruba,'Meaning':english})
```
